```python
import pandas as pd
import matplotlib.pyplot as plt

import warnings
warnings.filterwarnings('ignore')

%matplotlib inline
```

```python
# f, ax = plt.subplots(1)
# ax.grid(True, which='both')
# plt.tick_params(
#     axis='x',          # changes apply to the x-axis
#     which='both',      # both major and minor ticks are affected
#     bottom='off',      # ticks along the bottom edge are off
#     top='off',         # ticks along the top edge are off
#     labelbottom='off') # labels along the bottom edge are off

def plot_df(df, _kind, _delta, _title):
    if _kind=='scatter':
        _figsize=(8,4)
        ax = df.plot(kind=_kind, x=df.columns[0], y=df.columns[1], figsize=_figsize, title=_title)
        ax.set_xlim(xmin=0)
    else:
        _figsize=(10,6)
        ax = df.plot(kind=_kind, figsize=_figsize, title=_title)
    ax.set_ylim(ymin=0)
    ax.yaxis.grid(True)
    start, end = ax.get_ylim()
    ax.yaxis.set_ticks(pd.np.arange(start, end, _delta))
```

# Statistics Basics

```python
df = pd.DataFrame([[7,86],[8,70],[6,56], [5,56], 
                   [6,70], [7,80], [6.5,72], [8.5,91],
                   [6.5,81], [7,86]], 
                  columns=["Hours Slept", "Temporal Memory"])
plot_df(df, 'scatter', 5, "sleep-memory correlation")
```

# [stroop test](https://faculty.washington.edu/chudler/java/ready.html)

- congruent: 9.637 seconds.
- incongruent: 19.831 seconds.  

```python
_myscore=[9.637, 19.831]
df = pd.read_csv('stroopdata.csv', index_col=False)
df = df.append(pd.DataFrame([_myscore], columns=df.columns), ignore_index=True)
print(df.describe())
plot_df(df, 'bar', 2, "stroop effect - congruent vs ingcongruent")
plot_df(df.describe()[1:], 'bar', 5, "stroop effect - stats") #exclude row 'count' in describe() output
```
